```python
with open('netlist') as f:
    raw_netlist_all = f.read()
```

```python
lines = []
parsing = False
for line in raw_netlist_all.split('\n'):
    if line == 'ELABORATED BRANCHES:':
        break
    elif parsing:
        lines.append(line)
    elif line == 'SCOPES:':
        parsing = True
raw_netlist = '\n'.join(lines)
scopes, elab_nodes = raw_netlist.split('ELABORATED NODES:\n')
```

```python
class IvlModule:
    def __init__(self, name, xtype, ports=None):
        self.name = name
        self.xtype = xtype
        self.ports = ports or []
    
    def __repr__(self):
        return '<IvlModule: %s "%s" (%s port(s))>' % (self.xtype, self.name, len(self.ports))
```

```python
from enum import Enum


class IvlPortTypes(Enum):
    reg = 1
    wire = 2
    event = 3


class IvlPortDirections(Enum):
    input = 1
    output = 2
    
class IvlPort:
    def __init__(self, name, xtype, code_snippet=None, nets=None, direction=None):
        self.name = name
        self.xtype = xtype
        self.code_snippet = code_snippet
        self.nets = nets or set()
        self.direction = direction
        
    def add_net(self, net):
        self.nets.add(net)
    
    def remove_net(self, net):
        self.nets.remove(net)
    
    def __repr__(self):
        output = '<IvlPort: '

        if self.direction:
            output += '%s ' % self.direction.name

        output += '%s ' % self.xtype.name

        output += '"%s" ' % self.name

        if self.code_snippet:
            output += '@ %s ' % self.code_snippet

        output = output[:-1]
        output += '>'
        return output
```

```python
class IvlNet:
    def __init__(self, xid, name, ports=None):
        self.xid = xid
        self.name = name
        self.ports = ports or set()

    def add_port(self, port):
        self.ports.add(port)
        
    def remove_port(self, port):
        self.ports.remove(port)
        
    def has_port(self, port):
        return port in self.ports
    
    def __repr__(self):
        return '<IvlNet: "%s" (%s port(s))>' % (self.name, len(self.ports))


class IvlNetManager:
    def __init__(self):
        self.nets = {}
        
    def add_net_member(self, net_id, net_name, port):
        if net_id not in self.nets:
            self.nets[net_id] = IvlNet(net_id, net_name)
        net = self.nets[net_id]
        net.add_port(port)
        port.add_net(net)
        return net
    
    def get_net(self, net_id):
        return self.nets[net_id]
```

```python
def leading_spaces(line):
    return len(line) - len(line.lstrip(' '))
```

```python
def parse_module_data(lines, net_manager):
    ports = []
    port = None
    skip = True
    for line in lines:
        if leading_spaces(line) == 4:
            if not skip:
                ports.append(port)
                port = None
            else:
                skip = False
            line = line.lstrip(' ')
            if line.startswith('reg') or line.startswith('wire'):
                if line.startswith('reg'):
                    xtype = IvlPortTypes.reg
                else:
                    xtype = IvlPortTypes.wire
                name = line.split(': ')[1].split('[')[0]
                direction_raw = line.split('logic')[1].split('(eref')[0].strip(' ')
                if direction_raw == 'input':
                    direction = IvlPortDirections.input
                elif direction_raw == 'output':
                    direction = IvlPortDirections.output
                else:
                    direction = None
                port = IvlPort(name, xtype, direction=direction)
            elif line.startswith('event'):
                xtype = IvlPortTypes.event
                name = line.split('event ')[1].split(';')[0]
                snippet = line.split('// ')[1]
                port = IvlPort(name, xtype, code_snippet=snippet)
            else:
                skip = True
        elif leading_spaces(line) == 8:
            if port:
                net_id, net_name = line.split(': ')[1].split(' ')
                net_manager.add_net_member(net_id, net_name, port)
    if port:
        ports.append(port)
    return ports
```

```python
net_manager = IvlNetManager()

module_lines = []
modules_lines = []
first = True
for line in scopes.split('\n'):
    if leading_spaces(line) == 0:
        if not first:
            modules_lines.append(module_lines)
        else:
            first = False
        module_lines = []
    module_lines.append(line)

modules = []
for module_lines in modules_lines:
    module_meta = module_lines[0]
    module_data = module_lines[1:]
    name, supertype, module_type_raw, inst_type = module_meta.split(' ')
    module_type = module_type_raw.lstrip('<').rstrip('>')
    ports = parse_module_data(module_data, net_manager)
    module = IvlModule(name, module_type, ports)
    modules.append(module)

for module in modules:
    print(module)
    for port in module.ports:
        print('   ', port)
        for net in port.nets:
            print('       ', net)
```
