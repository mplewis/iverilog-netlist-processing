```python
import subprocess
from IPython.core.display import Image

def show_dot(content):
    with open('tmp.dot', 'w') as f:
        f.write(content)
    subprocess.check_call('dot -Tpng tmp.dot >tmp.png', shell=True)
    return Image(filename='tmp.png')
```

```python
dot = """
digraph test {
    node [shape=record];
    s1 [label="{a | b | c} | { \<thing\> test.b | { t | u | v } } | { d | e | f }"]
}
"""

show_dot(dot)
```

```python
template = """
digraph test {
    node [shape=record];
    s1 [label="%s"]
}
"""

def module_dot_label(xtype, name, left=None, right=None, bottom=None):
    left = left or []
    right = right or []
    bottom = bottom or []
    left = ['<%s>%s' % (x.replace(' ', '_'), x) for x in left]
    left_str = ' | '.join(left)
    right = ['<%s>%s' % (x.replace(' ', '_'), x) for x in right]
    right_str = ' | '.join(right)
    bottom = ['<%s>%s' % (x.replace(' ', '_'), x) for x in bottom]
    bottom_str = ' | '.join(bottom)
    output = ''
    if left:
        output += '{ %s } | ' % left_str
    output += '{ \<%s\> %s ' % (xtype, name)
    if bottom:
        output += '| { %s } ' % bottom_str
    output += '}'
    if right:
        output += ' | { % s }' % right_str
    return output

show_dot(template % module_dot_label('Robot', 'Wall-E', ['Clock', 'Data', 'Reset'], ['Motor Fwd', 'Lights', 'Buzzer'], ['Watchdog', 'Sleep', 'Warning']))
```

```python
template = """
digraph test {
    node [shape=record];
    ripple [label="%s"];
    bargraph [label="%s"];
    ripple:out -> bargraph:in;
}
"""

ripple = module_dot_label('ripple', 'r', ['clk'], ['out'])
bargraph = module_dot_label('bargraph', 'b', ['in'], ['out'])
show_dot(template % (ripple, bargraph))
```
